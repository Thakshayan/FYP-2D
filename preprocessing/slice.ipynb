```python
def save_to_json(data, path):
  with open(path, 'w') as fp:
    json.dump(data, fp)
```

```python
obj = pd.read_csv(DATA_DIR + 'info_df.csv')
meta = pd.read_pickle(DATA_DIR + 'meta_info_ProstateX-0000.pickle')
```

## 2D Preprocessing

```python
SLICED_DATA_DIR = ROOT_DIR + "dataset/PLD/sliced_2D/"
```

```python
if not os.path.exists(SLICED_DATA_DIR):
    os.makedirs(SLICED_DATA_DIR)
```

```python
paths = []
for id in obj['pid']:
    image1 = np.load(f'{DATA_DIR}/{id}_img.npy') 
    image2 = np.load(f'{DATA_DIR}/{id}_rois.npy')
    for i in range(image1.shape[0]):
      a = np.transpose(image1[i,:,:,0:1],(2,0,1))
      b = np.transpose(image1[i,:,:,5:6],(2,0,1))
      image_path = os.path.join(SLICED_DATA_DIR, str(id) + '_'+ str(i) + '_image.npy')
      label_path = os.path.join(SLICED_DATA_DIR, str(id) + '_'+ str(i) + '_label.npy')
      np.save(image_path, a)
      np.save(label_path, b)
      paths.append({"image":image_path, "label":label_path})
    print(id)

save_to_json( {"path": paths}, SLICED_DATA_DIR + 'config.json')
print("completed")
```
